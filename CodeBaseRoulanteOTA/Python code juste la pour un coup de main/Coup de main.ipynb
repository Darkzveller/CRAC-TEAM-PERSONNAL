```python
import matplotlib.pyplot as plt

# Paramètres de la simulation
vf = 600

# Initialisation des listes
x_values_1 = []
cons_hyp = []
hyp_recalc = []
phase1 = []
phase2 = []

# Calcul des données
for i in range(0, vf, 2):
    x_values_1.append(i)
    cons_hyp.append(vf)
    hyp_recalc.append(vf - i)
    phase1.append(vf/3)
    phase2.append(vf*2/3)

# Tracé du graphique
plt.figure()
plt.plot(x_values_1, hyp_recalc, label="hyp_recalc", linewidth=1)
plt.plot(x_values_1, cons_hyp, label="cons_hyp", linewidth=1)
plt.plot(x_values_1, phase1, label="phase1", linewidth=1)
plt.plot(x_values_1, phase2, label="phase2", linewidth=1)

plt.xlabel("Itération (i)")
plt.ylabel("Valeurs des consignes")
plt.title("Évolution des consignes en fonction de i")
plt.legend()
plt.grid()
plt.show()

print("Premier graphique affiché.")

print(vf/2)

```

```python
import matplotlib.pyplot as plt

# Paramètres de la simulation
vf = 600

# Initialisation des listes
x_values_1 = []
cons_hyp = []
hyp_recalc = []

# Calcul des données
for i in range(0, vf, 2):
    x_values_1.append(i)
    cons_hyp.append(vf)
    hyp_recalc.append(vf - i)

# Définition des points uniques pour phase1 et phase2
phase1_x = vf / 5
phase1_y = vf / 5  # Y peut être ajusté selon tes besoins

phase2_x = vf * 3 / 4
phase2_y = vf * 3 / 4  # Y peut être ajusté selon tes besoins

# Tracé du graphique
plt.figure()
plt.plot(x_values_1, hyp_recalc, label="hyp_recalc", linewidth=1)
plt.plot(x_values_1, cons_hyp, label="cons_hyp", linewidth=1)

# Ajout des points uniques
plt.scatter(phase1_x, phase1_y, color='red', label="Phase 1", zorder=3)
plt.scatter(phase2_x, phase2_y, color='blue', label="Phase 2", zorder=3)

plt.xlabel("Itération (i)")
plt.ylabel("Valeurs des consignes")
plt.title("Évolution des consignes en fonction de i")
plt.legend()
plt.grid()
plt.show()

print("Premier graphique affiché.")
print("phase1_x", vf - phase1_y)

print("phase2_y.",vf - phase2_y)
```
